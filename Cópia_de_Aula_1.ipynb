```python
# Importando a lib OpenCV
import cv2 as cv
```

```python
# Importando numpy
import numpy as np
```

```python
# Roda imagem e exibe os gráficos dela:
%matplotlib inline
```

```python
# PLotar = Exibir uma gráfico
from matplotlib import pyplot as plt
```

```python
# Printando a versão do Open cv2
print(f'OpenCV Versão {cv2.__version__}')
```

```python
# Toda vez que subir uma imagem pelo open cv ele usa o sistema inverso do rgb, o bgr, deixando a imagem para daltônico
image = cv.imread("/content/images/NATUREZA_1.jpg")
```

```python
# Convertendo BGR para RGB
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
# Retira o eixo de coluna e linha
plt.axis('off')
# Plotar imagem
plt.imshow(image)
```

```python
# Dimensões da imagem
print(f'Dimensões da imagem: {image.shape}')
```

```python
# Exibindo o numero de linhas:
print(f'Dimensões da linha: {image.shape[0]}')
```

```python
# Exibindo o numero de linhas:
print(f'Dimensões da coluna: {image.shape[1]}')
```

```python
# Exibindo a quantidade de camadas:
print(f'Quantidade de camadas: {image.shape[2]}')
```

```python
# Printando uma imagem: (Cada linha da matriz é um rgb segmentado...)
print(f'Matriz da imagem: \n {image}')
```

```python
# Pegando a cor de um pixel específico da imagem
(r, g, b) = image[50,50]
```

```python
# Printando as cores do pixel:
print(f'O pixel 50, 50 tem as seguintes cores:\n'
      f'Vermelho: {r}\nVerde: {g}\nAzul: {b}')
```

# Trabalhando com imagem em escala cinza

```python
# Trabalhar com imagens cinzas facilita o processamento da imagem:
# Preto 0 --------- 255 Branco
image3 = cv.imread('/content/images/NATUREZA_1.jpg', cv.IMREAD_GRAYSCALE)
# Precisamos deixar cinza:
image3 = cv.cvtColor(image3, cv.COLOR_BGR2RGB)
```

```python
# o cmap plota cinza tbm, caso o comando de cima não eja feito
plt.imshow(image3)
```

```python
# Alterar o tmanho da imagem:
print(f'Velhas dimensões da imagem: {image3.shape}')
image4 = cv.resize(image3, (600,400), cv.INTER_LINEAR)
print(f'Novas dimensões da imagem: {image4.shape}')
plt.imshow(image4)
plt.show()
```
